```python
import numpy as np
import pandas as pd
import cv2
import pytesseract
from glob import glob
import spacy
import re
import string
import warnings

warnings.filterwarnings('ignore')
```

```python

def cleanText(txt):
    whitespaces = string.whitespace
    punctuation ="!#$%&\'“”()*+:;<=>?[\\]^`{|}~"
    tableWhitespace = str.maketrans('', '', whitespaces)
    tablePunctuation = str.maketrans('', '', punctuation)
    text = str(txt).lower()
    removeWhitespaces = text.translate(tableWhitespace)
    removePunctuation = removeWhitespaces.translate(tablePunctuation)
    
    return str(removePunctuation)
```

```python
# Load NER model
model_ner = spacy.load('./output/model-best/')
```

```python
# Load image
image = cv2.imread('./data/6.jpg')

# etraxt data
tessData = pytesseract.image_to_data(image)
tessData

```

```python
#convert into DataFrame
tessList=list(map(lambda x:x.split('\t'), tessData.split('\n')))
df = pd.DataFrame(tessList[1:], columns=tessList[0])
df
```

```python
#clean text
df.dropna(inplace=True)
df['text']=df['text'].apply(cleanText)
df
```

```python
df_clean = df.query('text!="" ')
content = " ".join([w for w in df_clean['text']])
content
```

```python
#get_prediction from NER model
doc = model_ner(content)
from spacy import displacy
displacy.render(doc, style='ent')
```

```python
#Lesson 44. Combine the BIO information
bb_df['label'] = bb_df['label'].apply(lambda x: x[2:])
bb_df.head()
```

```python
#group the labels
class GroupGen:
    
    def __init__(self):
        self.id = 0
        self.text = ''
    
    def get_group(self, text):
        if self.text == text:
            return self.id
        else:
            self.id+=1
            self.text = text
            return self.id

grp_gen = GroupGen()
```

```python
bb_df['group'] = bb_df['label'].apply(grp_gen.get_group)
```

```python
bb_df
```

```python
# Lesson 45 Bounding box
# right and bottom bounding box
bb_df[['left', 'top','width', 'height']] = bb_df[['left', 'top', 'width', 'height']].astype(int)
bb_df['right'] = bb_df['left'] + bb_df['width']
bb_df['bottom'] = bb_df['top'] + bb_df['height']
```

```python
#tagging: groupby group
col_group = ['left', 'top', 'right', 'bottom', 'label', 'token', 'group']
group_tag_img = bb_df[col_group].groupby(by='group')
```

```python
img_tagging = group_tag_img.agg({
        'left':min,
        'right':max,
        'top':min,
        'bottom':max,
        'label': np.unique,
        'token':lambda x: " ".join(x)
        })
```

Parser

```python
def parser(text, label):
    if label =='PHONE':
        text = text.lower()
        text = re.sub(r'\D', '', text)
    elif label == 'EMAIL':
        text = text.lower()
        allow_special_characters = '@_.\-'
        text = re.sub(r'[^A-Za-z10-9{} ]'.format(allow_special_characters), '', text)
    elif label == 'WEB':
        text = text.lower()
        allow_special_characters = ':/.%#\-'
        text = re.sub(r'[^A-Za-z10-9{} ]'.format(allow_special_characters), '', text)
    elif label in ('NAME', 'DES'):
        text = text.lower()
        text = re.sub(r'[^a-z ]', '', text)
        text = text.title()
    elif label == 'ORG':
        text = text.lower()
        text = re.sub(r'[^a-z0-9 ]', '', text)
        text = text.title()
    return text
```

```python
info_array = dataframe_info[['token', 'label']].values
entities = dict(NAME=[], ORG=[], DES=[], PHONE=[], EMAIL=[], WEB=[])
previous = 'O'

for token, label in info_array:
    bio_tag = label[:1]
    label_tag = label[2:]
    # parse the token
    text = parser(token, label_tag)
    
    if bio_tag in ('B', 'I'):
        if previous !=label_tag:
            entities[label_tag].append(text)
        else:
            if bio_tag == "B":
                entities[label_tag].append(text)
            else:
                if label_tag in ("NAME", "ORG", "DES"):
                    entities[label_tag][-1] = entities[label_tag][-1]+" "+text
                else:
                    entities[label_tag][-1] = entities[label_tag][-1]+text
    previous = label_tag
    
            
```
